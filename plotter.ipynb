```python
from PIL import Image, ImageDraw

# Function to parse the input string and generate a list of (x, y, (r, g, b))
def parse_input(data_str):
    points = []
    data_list = data_str.split()
    for item in data_list:
        x, y, r, g, b = map(int, item.split('-'))
        points.append((x, y, (r, g, b)))
    return points

# Function to draw the matrix based on the parsed points with a black border
def draw_matrix(points, width, height):
    # Create a new blank image with white background
    img = Image.new('RGB', (width, height), 'black')
    draw = ImageDraw.Draw(img)

    # Define the size of each cell in the grid
    cell_size = 10

    # Draw each point with a black border
    for x, y, color in points:
        top_left = (x * cell_size, y * cell_size)
        bottom_right = ((x + 1) * cell_size, (y + 1) * cell_size)
        draw.rectangle([top_left, bottom_right], fill=color, outline='black')

    # Show the image
    img.show()

points = parse_input(data_str)
draw_matrix(points, width=320, height=320)
```
